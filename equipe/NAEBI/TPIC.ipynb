```python
amarelo = [1,4,3,0,0]
branco  = [8,33,18,3,3]
ind = [0,1,1,0,0]
pardo = [33,56,37,8,5]
preto =[9,20,6,1,1]
```

```python
axis = ['1','1 a 3','3 a 6','6 a 9', 'mais de 9']
```

```python
def perc(a):
    s = sum(a)
    return [i/s*100 for i in a]
a = perc(amarelo)
b = perc(branco)
i = perc(ind)
pa = perc(pardo)
pr = perc(preto)
```

```python
def grafico1(axis,a,title):
    plt.plot(axis,a,'o-')
    plt.xlabel("Renda")
    plt.ylabel("Distribuição (%)")
    plt.title(title)
    plt.savefig(title+".png")
def grafico2(axis,a,title):
    plt.bar(axis,a)
    plt.xlabel("Renda")
    plt.ylabel("Distribuição")
    plt.title(title)
    plt.savefig(title+"1.png")
```

```python
grafico1(axis,a, "Amarelo")
```

```python
grafico2(axis,amarelo, "Amarelo")
```

```python
grafico1(axis,b, "Branco")
```

```python
grafico2(axis,branco, "Branco")
```

```python
grafico1(axis,i, "Indígena")
```

```python
grafico2(axis,ind, "Indígena")
```

```python
grafico1(axis,pa,"Pardo")
```

```python
grafico2(axis,pardo,"Pardo")
```

```python
grafico1(axis,pr,"Preto")
```

```python
grafico2(axis,preto,"Preto")
```

```python
def grafico_bar(axis,a,title, xlabel,ylabel):
    plt.bar(axis,a)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.savefig(title+".png")
```
